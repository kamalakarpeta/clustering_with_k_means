Below is a simplified script for a K-Means clustering algorithm using the iris dataset from scikit-learn. This dataset does have predefined labels, but we will ignore them for this unsupervised learning task.

First, let's import the necessary libraries and load the dataset:

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans

iris = datasets.load_iris()
X = iris.data
```

Next, we will initialize the K-Means algorithm. For simplicity, let's assume we want to find 3 clusters, which is the actual number of classes in the iris dataset:

```python
kmeans = KMeans(n_clusters=3)
```

Now, fit the model to the data:

```python
kmeans.fit(X)
```

The K-Means algorithm will now have found the centers of 3 clusters in the data. We can obtain the coordinates of these centers, as well as the labels it assigned to each data point:

```python
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

Finally, let's visualize the results. We'll plot the data points and color them according to their assigned label. We'll also plot the cluster centers:

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=300, alpha=0.5)
plt.show()
```

This will produce a scatter plot of the data, with each point colored according to its assigned cluster. The centers of the clusters are marked in red.

Please note that this is a simplified example. In a real-world application, you would need to preprocess your data and possibly tune the parameters of the K-Means algorithm.
